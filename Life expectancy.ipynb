# Data science project calculating life expectancy

```python
#import models and scoring and encoders to format variables.
# Prediction
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix

# Classifiers

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

# Data encoding from string to dummy
from sklearn.preprocessing import LabelEncoder
```

```python
#check columns
df.columns
```

```python
#encode the string variables
lb = LabelEncoder()
df['Country'] = lb.fit_transform(df['Country'])
df['Status'] = lb.fit_transform(df['Status'])
```

```python
#make a new column called LE to see who lives over 50 and who doesnt
df['LE'] = df['Life expectancy '].apply(lambda x: 0 if x < 60 else 1)
```

```python
#fill empty 
y = df['Life expectancy '].fillna(0)
```

```python
#features we will use to see how they affect life expectancy
features = ['Country','Status','Alcohol','Hepatitis B','Diphtheria ',' HIV/AIDS',
           'Schooling','Polio',' BMI ']
```

```python
#fill na
X=df[features].fillna(0)
```

```python
#split the dataset 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```


# Linear Regression

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print('Coefficients :n', lm.coef_)
```

```python
lrpredictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,lrpredictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
# calculate these metrics by hand!
from sklearn import metrics
from sklearn.metrics import accuracy_score
import numpy as np

print('MAE:', metrics.mean_absolute_error(y_test, lrpredictions))
print('MSE:', metrics.mean_squared_error(y_test, lrpredictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, lrpredictions)))
```

```python
sns.distplot((y_test-lrpredictions),bins=50)
```

```python
coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```
